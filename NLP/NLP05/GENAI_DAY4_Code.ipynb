```python
with open("/content/drive/MyDrive//aiffel/d4/state_of_the_union.txt") as f:
    text = f.read()
```

```python
#len은 어떤 기준으로 chunk size를 잴 것인가?의 기준이 되는 함수입니다.
#chunk_overlap은 chunk의 앞뒤로 다른 chunk와 설정한 size까지 겹칠 수 있도록 설정하는 것입니다.
text_splitter = CharacterTextSplitter(separator="\n\n", chunk_size=1000, chunk_overlap=100, length_function = len,)
chunks = text_splitter.split_text(text)
```

```python
print(chunks[0])
```

```python
length = []
for chunk in chunks:
    length.append(len(chunk))

print(length)
```

```python
tokenizer = tiktoken.get_encoding("cl100k_base")

def tiktoken_len(text):
    tokens = tokenizer.encode(
        text
    )
    return len(tokens)
```

```python
tiktoken_length = []
for chunk in chunks:
    tiktoken_length.append(tiktoken_len(chunk))

print(length)
print(tiktoken_length)
```

```python
# 문서 임베딩 생성
texts = [
    "This is red apple.",
    "This is yellow banana.",
    "This is green lime.",
]

embeddings = []
for text in texts:
    response = client.embeddings.create(
        model="text-embedding-3-small",
        input=text
    )
    embeddings.append(response.data[0].embedding)

print(f"✅ 임베딩 완료: {len(embeddings)}개 문서")
print(f"첫번째 임베딩 차원: {len(embeddings[0])}")
```

```python

def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
# 쿼리 정의 (리스트 형태)
queries = [
    "This is yellow banana."
]

# 쿼리 임베딩 (여러 개도 한 번에 처리 가능)
response = client.embeddings.create(
    model="text-embedding-3-small",
    input=queries,
)

# 첫 번째(유일한) 쿼리 임베딩
e_query = response.data[0].embedding

print(cos_sim(embeddings[0], e_query))
print(cos_sim(embeddings[1], e_query))
print(cos_sim(embeddings[2], e_query))
```

```python
loader = PyPDFLoader("/content/drive/MyDrive/aiffel/d4/Demian.pdf")
pages = loader.load_and_split()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0, length_function = tiktoken_len)
docs = text_splitter.split_documents(pages)
```

```python
tokenizer = tiktoken.get_encoding("cl100k_base")

def tiktoken_len(text):
    tokens = tokenizer.encode(text)

    return len(tokens)
```

```python
loader = PyPDFLoader("/content/drive/MyDrive/aiffel/d4/Demian.pdf")
pages = loader.load_and_split()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=5000, chunk_overlap=50, length_function = tiktoken_len)
texts = text_splitter.split_documents(pages)
```

```python
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# HuggingFace
# 384차원
hf_embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
docsearch = Chroma.from_documents(
    texts,
    hf_embedding,
    persist_directory="./chroma_hf",
)
```

```python
# OpenAI
# 1536차원
embedding_model = OpenAIEmbeddings(model="text-embedding-3-small") # "text-embedding-3-large"
docsearch = Chroma.from_documents(texts, embedding_model)
```

```python
# QA
# llm_gemini = ChatGoogleGenerativeAI(model = "gemini-1.5-flash", temperature=0.0)

llm_openai = ChatOpenAI(
    model="gpt-4.1-mini",  # 또는 "gpt-4o-mini"
    temperature=0.0,
)
```

```python
qa = RetrievalQA.from_chain_type(llm_openai, chain_type="stuff",
                                 retriever=docsearch.as_retriever(
                                     search_type="mmr",
                                     search_kwargs={"k": 3, "fetch_k" : 10 }
                                 ))
```

```python
query = "how demian looks like"
result = qa(query)
```
