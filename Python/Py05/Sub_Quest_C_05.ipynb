```python
import random as r

class Account:

    account_count = 0

    def __init__(self, name, balance=0):
        self.bank = 'SC은행'
        self.account = self.make_account()
        self.name = name
        self.balance = balance

        self.deposit_count = 0
        self.deposit_hist = []
        self.withdraw_hist = []
        
        Account.account_count += 1

    @classmethod
    def get_account_num(self): 
        return Account.account_count

    def make_account(self) -> str:
        accnt = str(r.randint(0, 10**11 - 1)).zfill(11)
        frst  = accnt[:3]
        scnd = accnt[3:5]
        thrd  = accnt[5:]
        return frst + '-' + scnd + '-' + thrd

    def deposit(self, amount) -> bool:
        if amount < 1:
            return False

        self.balance += amount
        self.deposit_count += 1
        self.deposit_hist.append(f"입급: {amount}")

        if self.deposit_count == 5:
            interest = int(self.balance * 0.01)
            self.balance += interest
            self.deposit_count = 0
            self.deposit_hist.append(f"이자: {interest}")
        return True


    def withdraw(self, amount) -> bool:
        if 0 < amount <= self.balance:
            self.balance -= amount
            self.withdraw_hist.append(f"출금: {amount}")
            return True
        return False

    def get_balance(self) -> int:
        return self.balance


    def display_info(self):
        print("----- 계좌 정보 -----")
        print(f"은행명: {self.bank}, 계좌번호: {self.account}, 예금주명: {self.name}, 잔액: {self.balance:,}원")
        print("--------------------")


    def deposit_history(self):
        print(f"[입금 내역] ({self.name}, {self.account})")
        if not self.deposit_hist:
            print("입금 내역 없음")
        else:
            for row in self.deposit_hist:
                print(row)

    def withdraw_history(self):
        print(f"[출금 내역] ({self.name}, {self.account})")
        if not self.withdraw_hist:
            print("출금 내역 없음")
        else:
            for row in self.withdraw_hist:
                print(row)

def main():

    accounts = [
        Account("홍길동", 500000),
        Account("김철수", 12000000),
        Account("이영희", 9500000)
    ]

    print("==================")
    print(f"생성 계좌 수: {Account.get_account_num()}개")
    print("==================")

    print("\n")
    print("==================")
    print("전체 계좌 정보")
    for acc in accounts:
        acc.display_info()
    print("==================")

    print("\n")
    print("==================")
    print("잔고 10,000,000원 이상 계좌")
    for acc in accounts:
        if acc.get_balance() >= 10_000_000:
            acc.display_info()
    print("==================")

    print("\n")
    print("==================")
    print("입/출금")
    acc_frst = accounts[0]
    for i in range(5):
        acc_frst.deposit(100000)
    acc_frst.withdraw(50000)

    acc_frst.display_info()
    acc_frst.deposit_history()
    acc_frst.withdraw_history()


if __name__ == "__main__":
    main()
```

```python
import random


class Account:
    # Q2. 클래스 변수: 계좌 개수
    account_count = 0

    # Q1. 생성자
    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance

        # 계좌번호 랜덤 생성 (3-2-6)
        self.account_number = (
            f"{random.randint(100, 999)}-"
            f"{random.randint(10, 99)}-"
            f"{random.randint(100000, 999999)}"
        )

        # 입금 횟수 및 거래 내역
        self.deposit_count = 0
        self.deposit_log = []
        self.withdraw_log = []

        # 계좌 개수 증가
        Account.account_count += 1

    # Q3. 계좌 개수 출력 메서드
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4 & Q7. 입금 메서드 (5회마다 이자 1%)
    def deposit(self, amount):
        if amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
            return

        self.balance += amount
        self.deposit_count += 1
        self.deposit_log.append(amount)

        # 입금 5회마다 이자 지급
        if self.deposit_count % 5 == 0:
            interest = self.balance * 0.01
            self.balance += interest

    # Q5. 출금 메서드
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액이 부족합니다.")
            return

        self.balance -= amount
        self.withdraw_log.append(amount)

    # Q6. 정보 출력 메서드
    def display_info(self):
        print(f"은행이름: {self.bank}")
        print(f"예금주: {self.owner}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔고: {self.balance:,.0f}원")

    # Q10. 입금 내역 출력
    def deposit_history(self):
        print("입금 내역")
        for amount in self.deposit_log:
            print(f"{amount:,}원")

    # Q10. 출금 내역 출력
    def withdraw_history(self):
        print("출금 내역")
        for amount in self.withdraw_log:
            print(f"{amount:,}원")


# ===========================
# Q8. 객체 3개 이상 생성
# ===========================
accounts = [
    Account("파이썬", 1_500_000),
    Account("자바", 500_000),
    Account("C++", 2_000_000),
]

# ===========================
# Q9. 잔고 100만원 이상 고객 출력
# ===========================
print("▶ 잔고 100만원 이상 고객")
for acc in accounts:
    if acc.balance >= 1_000_000:
        acc.display_info()
        print("-" * 30)

# ===========================
# 입금 / 출금 테스트
# ===========================
test_account = accounts[0]

test_account.deposit(100_000)
test_account.deposit(100_000)
test_account.deposit(100_000)
test_account.deposit(100_000)
test_account.deposit(100_000)  # 5번째 입금 → 이자 지급

test_account.withdraw(200_000)

# ===========================
# 거래 내역 출력
# ===========================
print("\n▶ 거래 내역")
test_account.deposit_history()
test_account.withdraw_history()

# ===========================
# 전체 계좌 개수 출력
# ===========================
print(f"\n전체 계좌 개수: {Account.get_account_num()}")
```

```python
import random as r

class Character:
    def __init__(self, name, lev, hp, att, dfns):
        self.name = name
        self.lev = lev
        self.hp = hp
        self.att = att
        self.dfns = dfns

    def is_alive(self) -> bool:
        return self.hp > 0

    def take_damage(self, damage: int):
        calc_dmg = damage - self.dfns
        if calc_dmg > 0:
            self.hp -= calc_dmg
            if self.hp < 0:
                self.hp = 0
    
    def attack_target(self, target: "Character") -> int:
        damage = r.randint(1, self.att)
        target.take_damage(damage)
        return damage
    
    
class Player(Character):
    def __init__(self, name: str):
        super().__init__(name=name, lev=1, hp=100, att=25, dfns=5)
        self.exp = 0

    def gain_experience(self, amount):
        self.exp += amount

        if self.exp >= 50:
            self.exp -= 50
            self.level_up()

    def level_up(self):
        self.lev += 1
        self.att += 10
        self.dfns += 5

class Monster(Character):
    def __init__(self, name="mob", lev=1):
        
        super().__init__(
            name, 
            lev, 
            r.randint(10, 30) * lev, 
            r.randint(5, 20) * lev, 
            r.randint(1, 5) * lev
        )


def battle(player: Player, monster: Monster) -> bool:
    while player.is_alive() and monster.is_alive():

        if r.randint(1, 10) %2 == 0:
            monster.take_damage(player.att)
            print(f"{player.name}이 {monster.name}에게 {player.att}만큼 공격했다.")
            print(f"{monster.name}의 체력: {monster.hp}")
            
        else:
            player.take_damage(monster.att)
            print(f"{monster.name}이 {player.name}에게 {player.att}만큼 공격했다.")
            print(f"{player.name}의 체력: {player.hp}")


    if player.is_alive():
        print("전투 승리!/n고블린 을 이겼다!")
        gained_exp = monster.lev * 20
        print(f"경험치 {gained_exp} 획득!")
        player.gain_experience(gained_exp)
        return True
    else:
        print("전투 패배..")
        return False
    
def main():
    monster_dict = {"슬라임": 1, "고블린": 2, "오크": 3}

    name = input("플레이어 이름 입력: ")
    player = Player(name)

    for m_name, m_lev in monster_dict.items():
        monster = Monster(m_name, m_lev)
        alive = battle(player, monster)
        if not alive:
            print("게임오버")
            return
    print("게임 끝")

if __name__ == "__main__":
    main()
```

## GPT

```python
import random as r


class Character:
    def __init__(self, name: str, lev: int, hp: int, att: int, dfns: int):
        self.name = name
        self.lev = lev
        self.hp = hp
        self.att = att
        self.dfns = dfns

    # 현재 체력이 0 이상인지 반환
    def is_alive(self) -> bool:
        return self.hp > 0

    # (받은 데미지 - 방어력) 만큼 체력 감소 (음수면 0 처리)
    def take_damage(self, damage: int):
        calc_dmg = damage - self.dfns
        if calc_dmg > 0:
            self.hp -= calc_dmg
            if self.hp < 0:
                self.hp = 0

    # 1 ~ 공격력 사이의 랜덤 데미지를 타겟에게 입힘
    def attack_target(self, target: "Character") -> int:
        dmg = r.randint(1, self.att)
        target.take_damage(dmg)
        return dmg


class Player(Character):
    def __init__(self, name: str):
        # 레벨 1, 체력 100, 공격력 25, 방어력 5
        super().__init__(name=name, lev=1, hp=100, att=25, dfns=5)
        self.exp = 0  # 경험치

    # 경험치 획득
    def gain_experience(self, amount: int):
        self.exp += amount
        # 50 이상이면 레벨업 (여러 번 가능)
        while self.exp >= 50:
            self.exp -= 50
            self.level_up()

    # 레벨 +1, 공격력 +10, 방어력 +5
    def level_up(self):
        self.lev += 1
        self.att += 10
        self.dfns += 5
        print(f"[레벨업] {self.name} 레벨 {self.lev} 달성! "
              f"(공격력 {self.att}, 방어력 {self.dfns})")


class Monster(Character):
    def __init__(self, name: str = "Monster", lev: int = 1):
        # 체력: 10~30 * 레벨
        hp = r.randint(10, 30) * lev
        # 공격력: 5~20 * 레벨
        att = r.randint(5, 20) * lev
        # 방어력: 1~5 * 레벨
        dfns = r.randint(1, 5) * lev
        super().__init__(name=name, lev=lev, hp=hp, att=att, dfns=dfns)


def battle(player: Player, monster: Monster) -> bool:
    # Player와 Monster가 서로 공격하다가 둘 중 하나의 체력이 0 이하가 될 때까지 진행
    print(f"\n=== {player.name} vs {monster.name} 전투 시작! ===")
    print(f"{player.name} (Lv {player.lev}, HP {player.hp})")
    print(f"{monster.name} (Lv {monster.lev}, HP {monster.hp})\n")

    while player.is_alive() and monster.is_alive():
        # 플레이어 턴
        dmg = player.attack_target(monster)
        print(f"{player.name}이(가) {monster.name}에게 {dmg} 데미지를 입혔습니다. "
              f"({monster.name} HP: {monster.hp})")
        if not monster.is_alive():
            break

        # 몬스터 턴
        dmg = monster.attack_target(player)
        print(f"{monster.name}이(가) {player.name}에게 {dmg} 데미지를 입혔습니다. "
              f"({player.name} HP: {player.hp})")

    if player.is_alive():
        print("전투 승리!")
        gained_exp = monster.lev * 20
        print(f"경험치 {gained_exp}를 획득했습니다.")
        player.gain_experience(gained_exp)
        return True
    else:
        print("전투 패배..")
        return False


# 사용 예시
if __name__ == "__main__":
    name = input("플레이어 이름을 입력하세요: ")
    player = Player(name)
    monster_level = r.randint(1, 3)
    monster = Monster("슬라임", monster_level)
    battle(player, monster)
```
