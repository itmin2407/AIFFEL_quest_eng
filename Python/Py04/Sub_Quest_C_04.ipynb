```python
# 문제 3-3. 클래스 업데이트  
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요! 

class Kiosk:
    def __init__(self, menu, price):
        self.menu = menu
        self.price = price
        self.selection = []

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        while True:
            try:
                num = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력

                # 메뉴판에 있는 음료 번호일때
                if 1 <= num & num <= len(self.menu):
                    self.price_sum = self.price[num - 1]  # 선택 음료의 가격
                    break
                
                # 메뉴판에 없는 번호일 때
                else:  
                    print("없는 메뉴입니다. 다시 주문해 주세요.")

            except ValueError:
                print("숫자만 입력해 주세요.")
        
        while True:
            try:
                hot_or_ice = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                
                if hot_or_ice == 1:
                    self.temp = "HOT"
                    break
                elif hot_or_ice == 2:
                    self.temp = "ICE"
                    break
                else:    
                    print("1과 2 중 하나를 입력하세요.\n")
            except ValueError:
                print("숫자만 입력해 주세요.\n")

        print('주문 음료', self.temp, self.menu[num-1], ' : ', self.price_sum, '원')
        self.selection.append(self.menu[num-1])

    # 지불
    def pay(self):
        while True:
            try:
                choice = input("현금('cash' or 1), 카드('card' or 2)").strip()
                
                if choice in ['cash', '1']:
                    print("직원을 호출하겠습니다.")
                    break
                    
                elif choice in ['card', '2']:
                    print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                    break
                    
                else:
                    raise ValueError("다시 결제를 시도해 주세요.")
                    
            except ValueError as e:
                print(e)
                
    # 주문서 출력 
    def table(self):
        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        sum_price = 0
        # 주문 상품명 : 해당 금액
        for i in range(len(self.menu)):
            if self.menu[i] in self.selection:
                print(self.menu[i], ' : ', self.price[i])
                sum_price += self.price[i]
                
        print('합계 금액 :', sum_price)

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')

    def main(self):
        self.menu_print()
        self.menu_select()
        self.pay()
        self.table()

if __name__ == "__main__":
    menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
    price = [2000, 3000, 3000, 2500, 2500, 3000]
    kiosk = Kiosk(menu, price)
    kiosk.main()
```
